
In this notebook, we connect to the official ISTAT API to automatically fetch the latest available dataset on traffic accidents, defining a function for future API calls

```python
import requests
import pandas as pd

def fetch_istat_data(start_year):
#the url is generated from the official ISTAT API swagger
    url = f"https://esploradati.istat.it/SDMXWS/rest/data/41_983/all/?startPeriod={start_year}"
    header = {'Accept': 'application/vnd.sdmx.data+csv;version=1.0.0'}
    response = requests.get(url, headers=header)
        
    if response.status_code == 200:
        df1 = pd.read_csv(url, storage_options=header)
        return df1
    else:
        print(f"Error. Status code:{response.status_code}")
```

As requested, we fetch the actual data for the last 24 years

```python
df_traffic_incidents = fetch_istat_data(2023)
```
